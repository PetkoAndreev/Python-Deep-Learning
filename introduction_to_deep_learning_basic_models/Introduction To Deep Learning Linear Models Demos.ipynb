```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
```

<h1>Introduction To Deep Learning - Linear Models</h1>

```python
iris_data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", header = None)
```

```python
targets = iris_data[4]
```

```python
data = iris_data.drop(4, axis = 1)
```

```python
data
```

```python
data.describe().T
```

```python
targets = targets.replace({"Iris-setosa": 0, "Iris-versicolor": 1, "Iris-virginica": 2})
```

```python
tf.keras.backend.clear_session()
```

```python
# Dense - Definition of layer in NN. Each layer is related between prev and next layer.
logistic_regression = Sequential([
    Input(shape = (4,)),
    Dense(3, activation = "softmax", kernel_regularizer = "l2")
])
```

```python
logistic_regression.summary()
```

```python
logistic_regression.compile(
    optimizer = "adam",
    loss = "sparse_categorical_crossentropy", # Classification metric
    metrics = ["accuracy"]
)
```

```python
history = logistic_regression.fit(data, targets, epochs = 250)
```

```python
plt.plot(history.history["loss"])
plt.title("Learning curve", pad = 20)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
plt.plot(history.history["accuracy"])
plt.title("Learning curve", pad = 20)
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.show()
```

```python
actual = logistic_regression.predict(data)
```

```python
tf.reduce_mean(tf.keras.metrics.sparse_categorical_accuracy(targets, actual))
```

```python
logistic_regression.evaluate(data, targets)
```
